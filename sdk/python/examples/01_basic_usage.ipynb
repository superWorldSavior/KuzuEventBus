````python
import sys
import os

# Add parent directory to path for development
sys.path.insert(0, os.path.join(os.path.dirname(os.getcwd()), 'python'))

import casys_db
import tempfile
````

## 1. Initialize Engine and Database

````python
# Create a temporary directory for this example
tmpdir = tempfile.mkdtemp()
print(f"Using data directory: {tmpdir}")

# Initialize engine
engine = casys_db.CasysEngine(tmpdir)
print("✓ Engine initialized")
````

````python
# Open or create database
db_name = "social"
engine.open_database(db_name)
print(f"✓ Database '{db_name}' opened")

# Create main branch
try:
    engine.create_branch(db_name, "main")
    print("✓ Branch 'main' created")
except:
    print("✓ Branch 'main' already exists")

# Open branch
branch = engine.open_branch(db_name, "main")
print("✓ Branch opened")
````

## 2. Execute Raw GQL Queries

````python
# Execute a simple query
result = branch.query("MATCH (n:Person) RETURN n LIMIT 10")

print("Query result:")
print(f"  Columns: {result['columns']}")
print(f"  Rows: {result['rows']}")
print(f"  Count: {len(result['rows'])}")
````

## 3. Persistence Operations

````python
# Flush data to disk
branch.flush()
print("✓ Data flushed to disk")

# Load data from disk
branch.load()
print("✓ Data loaded from disk")
````

## 4. Create Nodes Directly

````python
# Add a node using low-level API
node_id = branch.add_node(["Person"], {"name": "Alice", "age": 30})
print(f"✓ Created node with ID: {node_id}")

# Query it back
result = branch.query("MATCH (p:Person) WHERE p.name = 'Alice' RETURN p.name, p.age")
print(f"Query result: {result}")
````

## 5. Cleanup

````python
# Clean up temporary directory
import shutil
shutil.rmtree(tmpdir)
print("✓ Cleaned up temporary directory")
````
